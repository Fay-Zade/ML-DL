```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
#For Jupyter Notebooks to show the plots
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.keys()
```

```python
data = iris['data']
type(data)
```

```python
df_iris = pd.DataFrame(data = iris['data'], columns=iris['feature_names'])
df_iris.head(5)
```

Here we choose Sepal Length @ column 0 and Petal Length @ column 2

```python
X = df_iris[['sepal length (cm)', 'petal width (cm)']]
X.head(5)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
for i, k in enumerate([2, 3]):
    fig, ax = plt.subplots(figsize=(10, 6))

    # Run the Kmeans algorithm
    km = KMeans(n_clusters=k)
    km.fit(X)
    labels = km.labels_
    centroids = km.cluster_centers_
#print(centroids)
    
    # Get silhouette samples
    silhouette_vals = silhouette_score(X, labels)
    #print(silhouette_vals)
    # scatter plot
    plt.scatter(X['sepal length (cm)'], X['petal width (cm)'], c=labels)    
    plt.scatter(centroids[:,0], centroids[:,1], marker='*', s=300, c='r', label='centroid')
    plt.xlabel('AnnualIncome')
    plt.ylabel('Loan Amount (In Thousands)')
    plt.title('Visualization of clustered data for k = {}'.format(k), fontweight='bold')
    plt.legend(loc =4)
    
```
